```python
import numpy as np
import librosa, librosa.display
import os
import tqdm
import time
import pdb
import matplotlib.pyplot as plt
import math
from skimage.feature import peak_local_max
import pickle
```

### The following cell contains the list of variables used to experiment with the parameters of the system. 

```python
hop_size = 512
fft_size = 1024
show_plots=False
use_cqt = False

# edit these to match the location and name of the recorded data
db_wav_dir = './sub_db'
q_wav_dir = './sub_q'

match_method_name = 'singlePeaksHop512Fft1024'
db_fprints_root = './database_fingerprints'
q_fprints_root = './query_fingerprints'


db_fprints_dir = os.path.join(db_fprints_root, match_method_name)
q_fprints_dir = os.path.join(q_fprints_root, match_method_name)
output_file_path = os.path.join(q_wav_dir, match_method_name +'_results.txt')
all_dirs = [db_fprints_dir, q_fprints_dir]

for d in all_dirs:
    if not os.path.exists(d):
        os.makedirs(d)
```

### In the following cell is a list of functions used to achieve each step in the document-retrieval process.

```python
"""
Recordings from the directory database_recordings are converted into spectrograms.
Peaks within a specified context (default setting ensures all peaks are at least 10 points away from one another)
in the spectrogram are identified using skimage's convenient peak_local_max function.
These are then converted to a list of coordinates.
Note that the word 'coordinates' is shortened to 'cord' in most cases for ease of programming.
"""

def audio_to_low_rez_cords(file_path, show_plots):
    mags_spec = get_magspec(file_path)
    coordinates = peak_local_max(np.log(mags_spec), min_distance=10,threshold_rel=None,indices = False)
    show_spec_peaks(mags_spec, coordinates, show_plots)
    cord_list = join_cords(coordinates)
    low_rez_cords = reduce_cord_dims(cord_list, mags_spec, 5)
    return low_rez_cords

# convert audio files to magnitude spectrograms
def get_magspec(file_path):
    # sample rates  of query and db wavs differ      
    x, Fs = librosa.load(file_path, sr=22050)
    if use_cqt == True:
        mags_spec = np.abs(librosa.cqt(x,window='hann',hop_length=hop_size))
    else:
        mags_spec = np.abs(librosa.stft(x,n_fft=fft_size,window='hann',win_length=fft_size,hop_length=hop_size))
    mags_spec[mags_spec==0] = 1e-40
    mags_spec = np.swapaxes(mags_spec,0,1) # because librosa swaps axes around for trad melspec representation
    return mags_spec

# converts a boolean array to a list of coordinates
def join_cords(coordinates):
    x_cords = np.where(coordinates == True)[0]
    y_cords = np.where(coordinates == True)[1]
    cord_list = []
    for i in range(len(x_cords)):
        cord_list.append((x_cords[i], y_cords[i]))
    return cord_list

# converts a list of coordinates to a hashed representation
def cord_list_to_hashed(cord_list):
    cord_list_by_bins = sorted(cord_list, key=lambda x: x[1])
    #+1 to include top index as bin
    cord_list_num_bin = cord_list_by_bins[-1][1] +1
    hashed_lists = [[] for i in range(cord_list_num_bin)]
    for peak_coord in cord_list:
            hashed_lists[peak_coord[1]].append(peak_coord[0])
    return hashed_lists

# condenses cords to a lower frequency resolution
# removes first n timesteps without peaks
def reduce_cord_dims(cord_list, mags_spec, hash_bin_size=5):
    num_freq_bins = mags_spec.shape[1]
    num_hash_bins = math.ceil(num_freq_bins/hash_bin_size)
    lower_dim_cords = []
    for hash_bin in range(num_hash_bins):
        start_freq = hash_bin*hash_bin_size
        hash_freq_range = range(start_freq,start_freq+hash_bin_size)
        for cord in cord_list:
            if cord[1] in hash_freq_range:
                lower_dim_cords.append((cord[0], hash_bin))
    lower_dim_cords = sorted(lower_dim_cords, key=lambda x: x[0])
    min_t_step = lower_dim_cords[0][0]
    lower_dim_cords = [(cord[0]-min_t_step, cord[1]) for cord in lower_dim_cords]
    return lower_dim_cords

# prints images of spectrograms and peaks
def show_spec_peaks(mags_spec, coordinates, show_plots):
    if show_plots == True:
        plt.figure(figsize=(20, 10))
        plt.imshow(librosa.amplitude_to_db(mags_spec,ref=np.max))
        plt.imshow(coordinates,cmap=plt.cm.gray_r, alpha=0.8, origin='lower')
        plt.show()
        plt.close()

# prints plotted graphs of an array
def show_plot(array):
    plt.figure(figsize=(20, 10))
    plt.plot(array)
    plt.show()
    plt.close()

# retrieves the minimum and maximum timestamp for a hashed list
def get_minmax_time(hashed_lists):
    peak_max_time = 0
    peak_min_time = 100000
    for hash_data in hashed_lists:
        if len(hash_data) > 0:
            if max(hash_data) > peak_max_time:
                peak_max_time = max(hash_data)
            if min(hash_data) < peak_min_time:
                peak_min_time = min(hash_data)
    return peak_min_time, peak_max_time

"""
The following function executes the procedure for generating constellation maps from a database's
audio documents and storing the corresponding hashed lists.
NOTE - This is not necessary to run if you already have populated fingerprint directories
(accompanied with this submission)!
"""

def fingerprintBuilder(wavs_path, fingerprints_path, show_plots=False):
    for file in os.scandir(wavs_path):
        if not file.name.endswith('.wav'):
            continue
        print('building fp for db file ', file.name)
        lower_dim_cords = audio_to_low_rez_cords(file.path, show_plots)
        hashed_lists = cord_list_to_hashed(lower_dim_cords)
        with open(os.path.join(fingerprints_path, file.name[:-4] +'.pkl'), 'wb') as handle:
            pickle.dump(hashed_lists, handle)
    print('Fingerprints generated!')

"""
The following functions define and executes the procedure for generating constellation maps from query recordings
and matching these to the audio fingerprints of the database documents.
The similarity scores for every comparison between queries and documents are retrieved.
The documents with the highest similarity score are given the highest rank in the prediction process.
"""
def single_peak_compare(lower_dim_cords, d_hashed_lists, show_plots):
    d_min_time, d_max_time = get_minmax_time(d_hashed_lists)
    d_length = d_max_time - d_min_time
    q_length = lower_dim_cords[-1][0] - lower_dim_cords[0][0] +1 #as we should be counting zeroth element too
    combined_length = d_length + q_length
    shifted_list = [[] for i in range(len(lower_dim_cords))]
    for i, coord in enumerate(lower_dim_cords):
        shift_n = coord[0]
        bin_idx = coord[1]
        try:
            shifted_list[i] = np.asarray(d_hashed_lists[bin_idx]) - shift_n
        except:
            continue
    sim_score = []
    binary_table = np.zeros((combined_length,len(shifted_list)))
    for i, shifted_seq in enumerate(shifted_list):
        if len(shifted_seq)==0:
            continue
        row = binary_table[:,i]
        row[shifted_seq] = 1
    sim_score = np.sum(binary_table, axis=1)
    if len(np.where(sim_score == np.max(sim_score))[0]) > 1:
        best_score_idx = int(np.where(sim_score == np.max(sim_score))[0][0])
    else: best_score_idx = int(np.where(sim_score == np.max(sim_score))[0])
    return sim_score, best_score_idx

def audioIndentification(q_wav_dir, db_fprints_dir, output_file_path, show_plots = False):  
    with open(output_file_path , 'w') as textFile:  
        for query_file in os.scandir(q_wav_dir):
            if not query_file.name.endswith('.wav'): continue
            print('building fp for query file', query_file.name)
            lower_dim_cords = audio_to_low_rez_cords(query_file.path, show_plots)
            db_scores = []
            for dbf_file in os.scandir(db_fprints_dir):
        #         if not dbf_file.name == 'pop.00097.pkl': continue
                if not dbf_file.name.endswith('.pkl'): continue
                print('comparing the two files: ', query_file.name, dbf_file.name)
                d_hashed_lists = pickle.load(open(dbf_file, 'rb'))
                sim_score, best_score_idx = single_peak_compare(lower_dim_cords, d_hashed_lists, show_plots)
                db_scores.append((sim_score[best_score_idx], dbf_file.name[:-4] +'.wav'))
            top_hits = sorted(db_scores, reverse=True)[:3]
            textFile.write(query_file.name + '\t' +top_hits[0][1] +'\t' +top_hits[1][1] +'\t' +top_hits[2][1] +'\n')
        textFile.write('time taken: ' +str(time.time()-start_time) +'\n')
        textFile.close()
        print('Predictions made!')
        
"""
Results of the audio identification process can be quantified in the following cell. For transparency,
accuracy is retrieved for the all queries, as well as the pop and classical music queries separately.
It can be deduced from the high accuracy scores pop music constellation maps are is considerably
more robust than those of classical music.
"""

def interpret_relevance_by_genre(output_file_path):
    genre = ['pop','classical']
    f = open(output_file_path , 'r')
    lines = f.readlines()
    mat_score = np.zeros((len(lines), 3))
    class_score = []
    pop_score = [] 
    for i, line in enumerate(lines):
        if line.startswith('time taken'): continue
        org = line.split('\t')[0]
        org_ident = org.split('-')[0]
#         query_score = None
        query_score = []
        for j, prediction in enumerate(line.split('\t')[1:]):
            if org_ident in prediction: score = 1
            else: score = 0    
            mat_score[i,j] = score
            query_score.append(score)
        if genre[0] in org_ident:
            pop_score.append(np.asarray(query_score))
        else:
            class_score.append(np.asarray(query_score))
    pop_score = np.asarray(pop_score)
    class_score = np.asarray(class_score)
    return mat_score, pop_score, class_score

def accuracy_by_genre_ranks(all_scores):
    score_names = ['all', 'pop', 'classical']
    for i, score in enumerate(all_scores):
        print(f'Accuracy for {score_names[i]} songs:')
        summed_scores = np.sum(score, axis=0)
        for j in range(3):
            j += 1
            print(f'Accuracy at Rank {j}: ' +str(np.sum(summed_scores[:j], axis=0)/score.shape[0]))
        print()
```

```python
fingerprintBuilder(db_wav_dir, db_fprints_dir, show_plots)
start_time = time.time()
audioIndentification(q_wav_dir, db_fprints_dir, output_file_path, show_plots)
all_scores = interpret_relevance_by_genre(output_file_path)
accuracy_by_genre_ranks(all_scores)
print('time taken: ', time.time()-start_time)
```
